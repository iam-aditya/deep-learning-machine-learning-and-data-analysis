```python
import pandas as pd

df = pd.read_excel('https://ibm.box.com/shared/static/lw190pt9zpy5bd1ptyg2aw15awomz9pu.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2
                      )
```

```python
print(df.dtypes)
print(df.shape)
df.head(5)
```

```python
def con(exp):
    if type(exp) ==int:
        return str(exp)
    else:
        return exp
df.columns=list(map(con,df.columns))    
df.columns=df.columns.astype(str)
```

```python
df['total']=df.sum(axis=1)
```

```python
df.drop(['Type','Coverage','REG','DEV','AREA'],axis=1,inplace=True)
```

```python
print(df.head(3))
```

```python
df.rename(columns={'OdName':'country','AreaName':'continent','RegName':'region','DevName':'dev'},inplace=True)
```
